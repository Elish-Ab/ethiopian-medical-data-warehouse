```python
# List of CSV file paths
csv_files = [
    "../scripts/DoctorsET_data.csv",
    "../scripts/CheMed123_data.csv",
    "../scripts/lobelia4cosmetics_data.csv",
    "../scripts/yetenaweg_data.csv",
    "../scripts/EAHCI_data.csv"
]

# Load and merge all CSV files into one DataFrame
df = pd.concat([pd.read_csv(file) for file in csv_files], ignore_index=True)

# Show first 10 rows
# Load all CSVs into DataFrames
dfs = [load_csv(file) for file in file_paths]

# Combine all DataFrames
df = pd.concat(dfs, ignore_index=True)

# Show first few rows
df.head(10)
```
